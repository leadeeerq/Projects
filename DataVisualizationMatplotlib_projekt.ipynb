##Wczytanie bibliotek

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

##Podstawowe wykresy

####Wykres liniowy

```python
plt.plot([1, 2, 3], [2, 3, 6])
```

```python
x = [1, 2, 3]
y = [2, 3, 6]

_ = plt.plot(x,y)
plt.title('Wykres f(x)=y')
plt.xlabel('Oś x')
plt.ylabel('Oś y')
plt.show()
```

```python
_ = plt.plot(x,y)
plt.title('FONT', fontdict={'fontname': 'sans-serif', 'fontsize': 20})
plt.xlabel('Oś x')
plt.ylabel('Oś y')

plt.show()
```

```python
_ = plt.plot(x,y)
plt.title('TICKS', fontdict={'fontname': 'sans-serif', 'fontsize': 20})
plt.xlabel('Oś x')
plt.ylabel('Oś y')

plt.xticks([0, 1, 2, 3])
plt.yticks([0, 2, 4, 4.5, 4.75, 6])

plt.show()
```

```python
_ = plt.plot(x,y, label='Zmienna x')

plt.title('LEGEND', fontdict={'fontname': 'sans-serif', 'fontsize': 20})
plt.xlabel('Oś x')
plt.ylabel('Oś y')

plt.xticks([0, 1, 2, 3])
plt.yticks([0, 2, 4, 4.5, 4.75, 6])

plt.legend()
```

```python
_ = plt.plot(x,y, label='Zmienna x', color = 'red', linewidth = 3, linestyle = '--', marker = '*', markersize = 15, markeredgecolor = 'yellow')

plt.title('COLORS', fontdict={'fontname': 'sans-serif', 'fontsize': 20})
plt.xlabel('Oś x')
plt.ylabel('Oś y')

plt.xticks([0, 1, 2, 3])
plt.yticks([0, 2, 4, 4.5, 4.75, 6])

plt.legend()
```

```python
_ = plt.plot(x, y, 'r*--', label='Zmienna x')

plt.title('SKRÓCONA NOTACJA', fontdict={'fontname': 'sans-serif', 'fontsize': 20})
plt.xlabel('Oś x')
plt.ylabel('Oś y')

plt.xticks([0, 1, 2, 3])
plt.yticks([0, 2, 4, 4.5, 4.75, 6])

plt.legend()
```

```python
x2 = np.arange(0, 5, 0.5)

_ = plt.plot(x, y, 'r*--', label='Zmienna x')
_ = plt.plot(x2[:6], x2[:6]**2, 'b*--', label = 'x2^2')
_ = plt.plot(x2[5:], x2[5:]**2, 'g*--', label = 'x2^2')

plt.title('x, x2', fontdict={'fontname': 'sans-serif', 'fontsize': 20})
plt.xlabel('Oś x')
plt.ylabel('Oś y')

plt.xticks([0, 1, 2, 3, 4, 5])
plt.yticks([0, 2, 4, 6, 20])

plt.legend()
```

```python
plt.figure(figsize = (8,3), dpi = 100)
_ = plt.plot(x, y, 'r*--', label='Zmienna x')
_ = plt.plot(x2[:6], x2[:6]**2, 'b*--', label = 'x2^2')
_ = plt.plot(x2[5:], x2[5:]**2, 'g*--', label = 'x2^2')

plt.title('ROZMIAR', fontdict={'fontname': 'sans-serif', 'fontsize': 20})
plt.xlabel('Oś x')
plt.ylabel('Oś y')

plt.xticks([0, 1, 2, 3, 4, 5])
plt.yticks([0, 2, 4, 6, 20])

plt.legend()
```

####Wykres słupkowy

```python
labels = ['A', 'B', 'C']
values = [1, 3, 1.5]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('O')

plt.figure(figsize = (3,2))

plt.show()
```

```python
bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
  bar.set_hatch(patterns.pop(0))
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')

plt.figure(figsize = (3,2))

plt.show()
```
