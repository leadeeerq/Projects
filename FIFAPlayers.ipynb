#### Relation between the price and the overall skills of the player

```python
type(df['Release Clause'][0])
```

```python
import re

number = r'\d+\.*\d*'
amounts = r'M|K'

word_re = '\d+\.*\d*(M|K)*'

def word_to_value(word):
  values_dict = {"m": 1000000,
                 "k": 1000,}
  return values_dict[word]

def parse_word_syntax(string):
  value_string = re.search(number, string).group()
  value = float(value_string.replace(',', '')) #Usunięcie przecinka ze stringa w celu konwersji na float
  word = re.search(amounts, string, flags = re.I).group().lower() #flags = re.I -> Ignorowanie wielkich liter w tysiącach, milionach itd.
  word_value = word_to_value(word)
  return value*word_value

def money_conversion(money):
  if money == 'NaN':
    return None

  if isinstance(money, list): #Sprawdzenie czy jest listą
    money = money[0]

  words_syntax = re.search(word_re, money, flags = re.I)

  if words_syntax:
    return parse_word_syntax(words_syntax.group())
  
  else:
    return None
```

```python
mon = '226M'
# re.search(number, string).group()

print(money_conversion(mon))
```

```python
df_clause = df.copy()
df_clause['Release Clause'] = [x.strip('€') if type(x) == str else x for x in df_clause['Release Clause']]
df_clause['Release Clause'] = [money_conversion(x) if type(x) == str else x for x in df_clause['Release Clause'] ]
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df_clause['Release Clause'])
```

```python
# df.dropna(subset=['Release Clause'])
df_clause = df_clause[df_clause['Release Clause'] != 'NaN']
```

#### Weight of the players

```python
#1 pound = 0.45359237 kilograms
```

```python
df.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in df.Weight]
# df.Weight = df.Weight * 0.45359237 # zrobić tylko raz :)
df.Weight = round(df.Weight, 1)
df.Weight.head()
```

```python
df.Weight.describe()
```

```python
plt.style.use('ggplot')

light = df.loc[df.Weight < 70].count()[0]
medium = df.loc[(df.Weight >= 70) & (df.Weight < 80)].count()[0]
heavy = df.loc[(df.Weight >80)].count()[0]

weights = [light, medium, heavy]
labels = ['<70 kg', '70-80 kg', '>80 kg']
explode = (0.1, 0.05, 0.1)

plt.title('Weight of the players')
plt.pie(weights, labels = labels, autopct='%.2f%%', explode=explode)
plt.show()
```

#### Age and the maximum potential skill growth - difference between actual overall skills and potential depending on the age

```python
df_pot = df.copy()
df_pot['Difference'] = df_pot['Potential'] - df_pot['Overall']
df_pot = df_pot.groupby(['Age']).mean(['Difference']).reset_index()
df_pot[['Overall', 'Potential', 'Difference', 'Age']].head()
```

```python
import seaborn as sns
sns.set

plt.title("Potential growth of the skills depending on player's age")
plt.plot(df_pot.Age, df_pot.Difference)
plt.xlabel('Age')
plt.ylabel('Potential growth')
plt.show()
```
