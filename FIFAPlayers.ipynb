####Preffered foot

```python
left = df.loc[df['Preferred Foot'] == 'Left'].count()[0]
right = df.loc[df['Preferred Foot'] == 'Right'].count()[0]
```

```python
plt.style.use('ggplot')
explode = (0.1, 0.1)
plt.pie([left, right], labels = ('Left', 'Right'), autopct = '%.2f%%', explode = explode)
plt.title('Foot preferences')
plt.show()
```

#### Weight of the players

```python
#1 pound = 0.45359237 kilograms
```

```python
df.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in df.Weight]
# df.Weight = df.Weight * 0.45359237 # zrobić tylko raz :)
df.Weight = round(df.Weight, 1)
df.Weight.head()
```

```python
df.Weight.describe()
```

```python
plt.style.use('ggplot')

light = df.loc[df.Weight < 70].count()[0]
medium = df.loc[(df.Weight >= 70) & (df.Weight < 80)].count()[0]
heavy = df.loc[(df.Weight >80)].count()[0]

weights = [light, medium, heavy]
labels = ['<70 kg', '70-80 kg', '>80 kg']
explode = (0.1, 0.05, 0.1)

plt.title('Weight of the players')
plt.pie(weights, labels = labels, autopct='%.2f%%', explode=explode)
plt.show()
```

#### Age and the maximum potential skill growth - difference between actual overall skills and potential depending on the age

```python
df_pot = df.copy()
df_pot['Difference'] = df_pot['Potential'] - df_pot['Overall']
df_pot = df_pot.groupby(['Age']).mean(['Difference']).reset_index()
df_pot[['Overall', 'Potential', 'Difference', 'Age']].head()
```

```python
import seaborn as sns
sns.set

plt.title("Potential growth of the skills depending on player's age")
plt.plot(df_pot.Age, df_pot.Difference)
plt.xlabel('Age')
plt.ylabel('Potential growth')
plt.show()
```

#### Best teams in the game

```python
df_teams = df.copy()
df_teams = df_teams.groupby(['Club']).mean(['Overall']).sort_values(['Overall'], ascending=False).reset_index()
df_teams[['Club', 'Overall', 'Potential', 'Age']].head()
```

```python
top_10 = df_teams[['Club','Overall']][:10]
top_10

plt.figure(figsize = (8,3))
plt.title('Average overall skills in top 10 teams')
plt.xticks(rotation=90)
plt.ylabel('Overall skills')
plt.plot(top_10['Club'], top_10['Overall'])
plt.show()
```

#### 3 best italian teams - boxplot

```python
Juve = df.loc[df.Club == 'Juventus']['Overall']
Napoli = df.loc[df.Club == 'Napoli']['Overall']
Inter = df.loc[df.Club == 'Inter']['Overall']

labels = ['Juventus', 'Napoli', 'Inter']

plt.style.use('default')
plt.figure(figsize = (8,4))
plt.ylabel('Overall skills')
plt.boxplot([Juve, Napoli, Inter], labels=labels)
plt.show()
```
