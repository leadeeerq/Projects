# Groupby

```python
import numpy as np
```

```python
df.developer.value_counts()
```

```python
# df.groupby('developer').aggregate({'average_user_rating': np.mean}).nlargest(n=5, columns = 'user_rating_count')
df.groupby('developer').aggregate({'average_user_rating': np.mean, 'user_rating_count': sum}).nlargest(n=5, columns = 'user_rating_count')
```

```python
df.groupby('primary_genre').aggregate({'price': np.mean})
```
