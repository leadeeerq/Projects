####Utworzenie listy unikalnych typ√≥w

```python
type1 = df['Type 1'].unique().tolist()
type2 = df['Type 2'].unique().tolist()

in_first = set(type1)
in_second = set(type2)

in_second_but_not_in_first = in_second - in_first

types = type1 + list(in_second_but_not_in_first)
# types = types.dropna()
types = types[0:-1] #Usuniƒôcie nan

for typ in types:
    print(typ)
```
