```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head(5)
```

```python
df.info()
```

```python
df.describe().T
```

```python
df['Average'] = df.tip / df.total_bill *100
df.head(5)
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip')
```

```python
sns.set(font_scale = 1.2)
_ = sns.relplot(data = df, x = 'total_bill', y = 'tip')
```

```python
_ = sns.relplot(data = df, x = 'total_bill', y = 'tip', size = 'size')
```

```python
_ = sns.relplot(data = df, x = 'total_bill', y = 'tip', size = 'size', hue = 'size')
```

## Wiele wykresw

```python
# Wywietlenie kilku wykresw na podstawie zmiennych
_ = sns.relplot(data = df, x = 'total_bill', y = 'tip', size = 'size', col = 'time') 
```

```python
_ = sns.relplot(data = df, x = 'total_bill', y = 'tip', size = 'size', row = 'time') 
```

```python
_ = sns.relplot(data = df, x = 'total_bill', y = 'tip', size = 'size', col = 'time', row = 'smoker') 
```
